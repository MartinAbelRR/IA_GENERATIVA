Podemos determinar qué tan similar es una palabra a otras palabras en nuestro marco de datos después de tener un vector que represente esa palabra. Calculando la similitud del coseno del vector de palabras de nuestro término de búsqueda con cada palabra incrustada en nuestro marco de datos.

```python
import numpy as np

def distancia_coseno(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
import ast

# Convertir las incrustaciones de cadenas a listas de flotantes
# df['incrustaciones'] = df['incrustaciones'].apply(ast.literal_eval)

# Calcular la similitud de coseno entre las incrustaciones de las palabras
df['puntuacion_similitud'] = df['incrustaciones'].apply(lambda x: distancia_coseno(x, incrustaciones_texto))
```
